#### Exercise: Add query functions / TASK 3

```python
task3_query = """
    SELECT customers.CustomerName, customers.ContactNumber, subq.Cost 
    FROM customers
    JOIN bookings 
        ON customers.CustomersID = bookings.CustomerID
    JOIN orders 
        ON bookings.BookingID = orders.BookingID
    JOIN (SELECT OrderID, MenuID, SUM(TotalCost) AS Cost FROM order_menus GROUP BY OrderID, MenuID) AS subq 
        ON orders.OrderID = subq.OrderID
    WHERE subq.Cost > 60;
"""
cursor.execute(task3_query)
result = cursor.fetchall()
for rec in result:
    print(rec)
```
